```python
import matplotlib.pyplot as plt
import numpy as np
arrX = np.linspace(0,1,100)
print(arrX)
def expX(x):
    return np.exp(x)
```

```python
y = expX(arrX)
print(y)
fig = plt.figure(figsize=(8,6))
plt.plot(arrX, y, ".")
plt.xlabel("Time[milliseconds]")
plt.ylabel("Awesomeness")
plt.savefig("exp.pdf", facecolor="white", dpi=600, bbox_inches="tight")
```
